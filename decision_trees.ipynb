## Loading dataset

```python
df = pd.read_csv('/mnist.csv')
df = df.set_index('id')
df.head(3)
```

## Spliting Dataset train and test datasets

```python
x = df.drop(['class'], axis=1)
y =df['class']

x_train, x_test, y_train, y_test, = sklearn.model_selection.train_test_split(x,y)

print('df', df.shape)
print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

## Training model

```python
model = sklearn.tree.DecisionTreeClassifier()
model.fit(x_train, y_train)

```

## Testing the model

```python
y_predicted = model.predict(x_test)
accuracy = sklearn.metrics.accuracy_score(y_test, y_predicted)
accuracy
```

## Hiperparameter tunning:

```python
min_samples_split = range(2, 10)

criterion_list = ['gini', 'entropy']

result_df = pd.DataFrame(columns=['min_samples_split', 'Criterion', 'Accuracy'])

for i in min_samples_split:
    for a in criterion_list:
        model = sklearn.tree.DecisionTreeClassifier(random_state=i, criterion = a)
        model.fit(x_train, y_train)
        y_predicted = model.predict(x_test)
        accuracy = sklearn.metrics.accuracy_score(y_test, y_predicted)
        result_df = result_df.append({'min_samples_split':i, "Accuracy": accuracy, "Criterion": a}, ignore_index=True)
result_df
```

## Visualization Accuracy by sample split

```python
k_df = result_df[result_df['Criterion']=='gini']

fig = px.line(x=k_df["min_samples_split"], y=k_df["Accuracy"], labels={'x':'Min_samples_split', 'y':'Accuracy'})

fig.show()
```

## Visualization: Accuracy by Criterion only

```python
criterion_df = result_df[result_df['min_samples_split']==3]

fig = px.bar(criterion_df, x='Criterion', y='Accuracy')
fig.show()
```
